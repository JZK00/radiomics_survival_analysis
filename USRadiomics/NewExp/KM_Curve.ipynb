```python
import pandas as pd
import numpy as np
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt
from lifelines.statistics import logrank_test

# 读取数据
df = pd.read_csv('TNBC_with_2yDFS.csv')

# 创建 KaplanMeierFitter 对象
kmf = KaplanMeierFitter()

# 创建一个颜色映射
colors = {'High Risk': 'red', 'Low Risk': 'blue'}

# 设置图形
plt.figure(figsize=(10, 6))

# 初始化 logrank p 值计算的变量
p_value = None

# 分组处理：分别画出高风险和低风险组的 KM 曲线（标签已置换）
for risk_group in ['High Risk', 'Low Risk']:
    group_data = df[df['2yDFSRisk'] == risk_group]
    
    # 使用 KaplanMeierFitter 进行拟合
    kmf.fit(group_data['DFS_time'], event_observed=group_data['DFS'], label=risk_group)
    
    # 绘制 KM 曲线
    kmf.plot_survival_function(ci_show=True, color=colors[risk_group], linestyle='-' if risk_group == 'High Risk' else '--')

# 计算 log-rank test P值
high_risk_data = df[df['2yDFSRisk'] == 'High Risk']
low_risk_data = df[df['2yDFSRisk'] == 'Low Risk']

results = logrank_test(high_risk_data['DFS_time'], low_risk_data['DFS_time'], event_observed_A=high_risk_data['DFS'], event_observed_B=low_risk_data['DFS'])
p_value = results.p_value

# 添加标题和标签
#plt.title('Kaplan-Meier Curve for 2-Year OS Risk Group (After Swapping)', fontsize=16)
plt.xlabel('Time (Month)', fontsize=14)  # 横坐标改为 "Time (Month)"
plt.ylabel('Overall Survival', fontsize=14)  # 纵坐标改为 "Overall Survival"

# 添加横纵坐标的虚线，颜色为淡灰色，透明度低
for y in np.arange(0, 1.1, 0.2):
    plt.axhline(y=y, color='gray', linestyle='--', linewidth=1, alpha=0.3)  # 透明度设置为0.3

for x in np.arange(20, 121, 20):
    plt.axvline(x=x, color='gray', linestyle='--', linewidth=1, alpha=0.3)  # 透明度设置为0.3

# 添加P值标签
plt.text(50, 0.8, f'P-value = {p_value:.4f}', fontsize=14, color='black')

# 设置图例位置为右上角
plt.legend(title='Risk Group', fontsize=12, loc='upper right')

# 显示图形
plt.show()
```
